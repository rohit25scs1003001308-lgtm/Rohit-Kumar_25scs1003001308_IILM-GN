```python
df['Actors'] = df[['Star1', 'Star2', 'Star3', 'Star4']].agg(','.join, axis=1).apply(lambda x: x.split(',')[:4])
df['Director'] = df['Director'].apply(lambda x: x.split(','))
```

```python
def clean(sentence):
    temp = []
    for word in sentence:
        temp.append(word.lower().replace(' ', ''))
    return temp
```

```python
df['Genre'] = [clean(x) for x in df['Genre']]
df['Actors'] = [clean(x) for x in df['Actors']]
df['Director'] = [clean(x) for x in df['Director']]
```

```python
# combining all the columns data
columns = ['clean_overview', 'Genre', 'Actors', 'Director']
l = []
for i in range(len(df)):
    words = ''
    for col in columns:
        words += ' '.join(df[col][i]) + ' '
    l.append(words)
df['clean_input'] = l
df = df[['Series_Title', 'clean_input']]
df.head()
```

# Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df['clean_input'])
```

```python
# create cosine similarity matrix
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(features, features)
print(cosine_sim)
```

# Movie Recommendation

```python
index = pd.Series(df['Series_Title'])
index.head()
```

```python
def recommend_movies(title):
    movies = []
    if title not in index.values:
        print(f"Movie with title '{title}' not found in the dataset.")
        return movies
    idx = index[index == title].index[0]
    # print(idx)
    score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top10 = list(score.iloc[1:11].index)
    # print(top10)

    for i in top10:
        movies.append(df['Series_Title'][i])
    return movies
```

```python
recommend_movies('The Dark Knight Rises')
```
